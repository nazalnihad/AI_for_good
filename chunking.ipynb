```python
import pdfplumber
import pandas as pd
import re

def fill_none_values(rows):
  filled_rows = []
  current_values = [None] * len(rows[0])

  for row in rows:
      # print(row)
      for i, cell in enumerate(row):
          if cell is not None:
              current_values[i] = cell.replace('\n', ' ') # Replace '\n' with a space
          else:
              row[i] = current_values[i]

      # Only append the row if it contains a word
      if any(isinstance(cell, str) and re.search(r'\b\w+\b', cell) for cell in row):
          filled_rows.append(row)

  # Only append the table if it contains more than one row of meaningful data
  if len(filled_rows) > 1:
      return filled_rows
  else:
      return []

def extract_tables(pdf_path):
    all_tables = []

    with pdfplumber.open(pdf_path) as pdf:
        for page_number in range(len(pdf.pages)):
            page = pdf.pages[page_number]

            # Adjust table extraction settings
            table_settings = {  # Use text-based horizontal lines
                 "vertical_strategy": "lines_strict",
    "horizontal_strategy": "lines_strict" # Adjust tolerance for intersection detection
                                }

            tables = page.extract_tables(table_settings=table_settings)

            if tables:
                for table in tables:
                    filled_table = fill_none_values(table[1:])
                    df = pd.DataFrame(filled_table, columns=table[0])
                    all_tables.append(df)

    return all_tables
```

```python
from langchain.document_loaders import PDFPlumberLoader
import pandas as pd

def extract_tables_and_text(pdf_path):
   loader = PDFPlumberLoader(pdf_path)
   data = loader.load()
   tables = extract_tables(pdf_path)
   all_tables = []
   all_text = []

   for chunk in data:
       # Extract tables
       if isinstance(chunk, dict) and 'table' in chunk:
           df = pd.DataFrame(chunk['table'][1:], columns=chunk['table'][0])
           all_tables.append({'table': df, 'metadata': chunk['metadata']})

       # Extract text
       elif isinstance(chunk, dict) and 'text' in chunk:
           all_text.append({'text': chunk['text'], 'metadata': chunk['metadata']})

   return all_tables, all_text

if __name__ == "__main__":
   pdf_path = "farm.pdf" # Replace with the path to your PDF file
   extracted_tables, extracted_text = extract_tables_and_text(pdf_path)
   print(extracted_tables)
   for i, table in enumerate(extracted_tables):
       print(f"Table {i + 1}:")
       print(table['table'])
       print("Metadata:")
       print(table['metadata'])
       print("-" * 30)

   for i, text in enumerate(extracted_text):
       print(f"Text {i + 1}:")
       print(text['text'])
       print("Metadata:")
       print(text['metadata'])
       print("-" * 30)
```

```python
import os
import pdfplumber

def chunk_text(pdf_name, text, chunk_size, overlap, chunk_type, page_number):
   chunks = []
   index = 0
   chunk_index = 0
   while index < len(text):
       chunk = text[index : index + chunk_size]
       attrs = {
           "pdf_name": pdf_name,
           "page_number": page_number,
           "chunk_index": chunk_index,
           "chunk_type": chunk_type,
       }
       chunks.append((chunk, attrs))
       index += chunk_size - overlap
       chunk_index += 1
   return chunks

def extract_text_chunks(pdf_path, chunk_size, overlap, chunk_type):
   all_chunks = []
   pdf_name = os.path.basename(pdf_path)

   with pdfplumber.open(pdf_path) as pdf:
       for page_number in range(len(pdf.pages)):
           page = pdf.pages[page_number]
           text = page.extract_text()
           chunks = chunk_text(pdf_name, text, chunk_size, overlap, chunk_type, page_number)
           all_chunks.extend(chunks)

   return all_chunks
```

```python
def add_metadata(data, metadata):
 if isinstance(data, pd.DataFrame):
    for key, value in metadata.items():
        data.attrs[key] = value
 elif isinstance(data, tuple):
    data = (data[0], {**data[1], **metadata})
 return data

def extract_tables(pdf_path):
 all_tables = []
 pdf_name = os.path.basename(pdf_path)

 with pdfplumber.open(pdf_path) as pdf:
    for page_number in range(len(pdf.pages)):
        page = pdf.pages[page_number]

        # table extraction settings
        # table_settings = {
        #     "vertical_strategy": "lines_strict",
        # }

        tables = page.extract_tables()

        if tables:
            for i, table in enumerate(tables):
              filled_table = fill_none_values(table[1:])
              df = pd.DataFrame(filled_table, columns=table[0])
              metadata = {'pdf_name': pdf_name, 'page_number': page_number, 'chunk_type': 'table', 'chunk_index': i}
              df = add_metadata(df, metadata)
              all_tables.append(df)

 return all_tables

def extract_text_chunks(pdf_path, chunk_size, overlap, chunk_type):
 all_chunks = []
 pdf_name = os.path.basename(pdf_path)

 with pdfplumber.open(pdf_path) as pdf:
    for page_number in range(len(pdf.pages)):
        page = pdf.pages[page_number]
        text = page.extract_text()
        chunks = chunk_text(pdf_name, text, chunk_size, overlap, chunk_type, page_number)
        for chunk_index, chunk in enumerate(chunks):
            metadata = {'pdf_name': pdf_name, 'page_number': page_number, 'chunk_type': chunk_type, 'chunk_index': chunk_index}
            chunk = add_metadata(chunk, metadata)
            all_chunks.append(chunk)

 return all_chunks
```

```python
def extract_all_data(pdf_path, chunk_size, overlap, chunk_type):
   all_tables = extract_tables(pdf_path)
   all_chunks = extract_text_chunks(pdf_path, chunk_size, overlap, chunk_type)

   all_data = all_tables + all_chunks

   return all_data
```
