```python
def chunk_by_title_with_overlap(elements, overlap):
   chunks = chunk_by_title(elements)
   overlapped_chunks = []

   for i in range(len(chunks) - 1):
       chunks[i].text += chunks[i + 1].text[:overlap]
       chunks[i + 1].text = chunks[i + 1].text[overlap:]
       overlapped_chunks.append(chunks[i])
      #  print(chunks[i])

   overlapped_chunks.append(chunks[-1])
   return overlapped_chunks

chunk_by_title_with_overlap(elements,100)
```
