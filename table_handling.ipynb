```python
# import tabula
import camelot

def extract_tables_from_pdf(pdf_path):
    # Extract tables using tabula-py
#     tables_tabula = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True, stream=True, silent=True)

    # Extract tables using camelot
    tables_camelot = camelot.read_pdf(pdf_path, flavor='stream', pages='all', table_ares = {'x': 310, 'y': 110, 'w': 535, 'h': 196})

    # Display or process the extracted tables from tabula-py
#     for idx, table in enumerate(tables_tabula):
#         print(f"Table {idx + 1} (from tabula-py):")
#         print(table)
#         print("\n" + "=" * 50 + "\n")

    # Display or process the extracted tables from camelot
    for idx, table in enumerate(tables_camelot):
        print(f"Table {idx + 1} (from camelot):")
        print(table.df)
        print("\n" + "=" * 50 + "\n")

if __name__ == "__main__":
    pdf_path = "D:/projects/AI_for_good/git_test.pdf"
    extract_tables_from_pdf(pdf_path)
```
